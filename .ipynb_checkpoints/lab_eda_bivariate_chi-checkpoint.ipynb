## Part 2: Exploring Product Prices and Ratings Across Categories and Brands

### Objective

- Use the IQR (Interquartile Range) method.
- Products priced below `Q1 - 1.5 * IQR` or above `Q3 + 1.5 * IQR` will be considered outliers and removed from the dataset.
- The next steps will be done with the dataframe without outliers.
- Hint: you can check the last Check For Understanding at the end of the lesson *EDA Bivariate Analysis* for a hint on how to do this.

### Violin Plots

- Which product category tends to have the highest median price?
  - Don't filter here by top categories.

### Bar Charts

  - Don't filter here by top categories.

### Box Plots
